```python
sfo_train = sfo_train[~null_mask]
```

### Fix IATA codes and Airline Names

```python
belair_iata = lambda x: '4T' if x == 'BBB' else x
sfo_train.loc[:, 'operating_airline_iata_code'] = sfo_train.loc[:, 'operating_airline_iata_code'].map(belair_iata)
sfo_train.loc[:, 'published_airline_iata_code'] = sfo_train.loc[:, 'published_airline_iata_code'].map(belair_iata)
```

```python
for name, df in sfo_train.groupby('operating_airline_iata_code'):
    uniques = df['operating_airline'].unique()
    print(
        name,
        uniques,
        '***' if len(uniques) > 1 else ''
    )
```

```python
for name, df in sfo_train.groupby('published_airline_iata_code'):
    uniques = df['published_airline'].unique()
    print(
        name,
        uniques,
        '***' if len(uniques) > 1 else ''
    )
```

```python
united = lambda x: 'United Airlines' if x == 'United Airlines - Pre 07/01/2013' else x
sfo_train.loc[:, 'operating_airline'] = sfo_train.loc[:, 'operating_airline'].map(united)
sfo_train.loc[:, 'published_airline'] = sfo_train.loc[:, 'published_airline'].map(united)


jazz = lambda x: 'Jazz Aviation' if x == 'Air Canada Jazz' else x
sfo_train.loc[:, 'operating_airline'] = sfo_train.loc[:, 'operating_airline'].map(jazz)
sfo_train.loc[:, 'published_airline'] = sfo_train.loc[:, 'published_airline'].map(jazz)
```

### Fix Passenger Count dtype

```python
sfo_train['passenger_count'] = sfo_train['passenger_count'].astype(int)
```

### Format Time

```python
sfo_train['activity_period'] = (pd.to_datetime(sfo_train['activity_period'], format='%Y%m')
                                  .dt.to_period('M'))
```

### Dummy Variables

There is a 1-1 mapping for IATA and airline names now, so I'll drop airline names to simplify things. I will keep a mapping in a different dataframe for later reference if needed.

```python
sfo_train = sfo_train.rename(columns={
    'operating_airline_iata_code': 'operating_iata',
    'published_airline_iata_code': 'published_iata',
})
```

```python
operating = ['operating_iata', 'operating_airline']
published = ['published_iata', 'published_airline']
new = ['iata', 'airline']

iata_to_airline = (
    pd.concat([
        sfo_train[operating]
            .rename(columns=dict(zip(operating, new))),
        sfo_train[published]
            .rename(columns=dict(zip(published, new))),
    ])
    .drop_duplicates()
    .set_index('iata')
    .sort_index()
).squeeze()

iata_to_airline
```

```python
sfo_train = sfo_train.drop(columns=['operating_airline', 'published_airline'])
```

```python
categorical = [col for col, dtype in dict(sfo_train.dtypes).items() if dtype == np.dtype(object)]
sfo_train[categorical] = sfo_train[categorical].astype('category')
sfo_train.dtypes
```
